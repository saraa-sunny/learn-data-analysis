## üõ† Handle Missing Data

```python
# count missing values per column
df.isnull().sum()
```

```python
# count non-missing
df.notnull().sum() 
```

```python
# drop rows with any missing values
df.dropna()             
```

```python
 # drop rows only if these columns have missing values
df.dropna(subset=['Gender', 'Age'])
```

```python
# replace all missing values with 0
df.fillna(0) 
```

```python
# column-specific fill
df['Age']=df['Age'].fillna(df['Age'].median())
```

## üõ† Fix Column Names

```python
# remove extra spaces
df.columns = df.columns.str.strip()    
```

```python
# to lowercase
df.columns = df.columns.str.lower()   
```

```python
# replace spaces with underscores
df.columns = df.columns.str.replace(' ', '_') 
```

```python
# rename columnn name
df.rename(columns={'family_history_with_overweight': 'family_history'}, inplace=True)
```

```python
# display the first 5 rows - to check for applied changes when needed
df.head()
```

## üõ† Handle Duplicates

```python
# count duplicates
df.duplicated().sum()  
```

```python
# Show duplicates
df[df.duplicated()]    
```

```python
# remove the duplicates
df.drop_duplicates(inplace=True)
```

## Convert Data Types

```python
df['age'] = df['age'].astype(int)
```

```python
df['height']=df['height'].astype(float)
```

## Handle Categorical Data

```python
# to find categorical columns 
df.select_dtypes(include='object')
```

```python
# Encode manually
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
df.head()
```

## üõ† Text Cleaning

```python
df['mtrans'] = df['mtrans'].str.lower().str.strip()
df.head()
```

```python
df['nobeyesdad'] = df['nobeyesdad'].str.replace('[^a-zA-Z ]', '', regex=True)
df.head()
```

## üõ† Sort Data

```python
# Sort by one column
df.sort_values('weight', ascending=False,inplace=True)
df.head()
```

```python
# Sort by multiple columns
df.sort_values(['age', 'height'], ascending=[True, False])
```

```python
# Sort by index
df.sort_index(inplace=True)
df.head()
```

## üõ† Filter Data (Row Selection)

```python
# Single condition
df[df['age'] > 50]
```

```python
# Multiple conditions
df[(df['age'] > 50) & (df['weight'] > 65)]
```

##  üõ†Replace/Modify Values

```python
df.replace({'yes': 1, 'no': 0}, inplace=True)
```

```python
df['smoke'] = df['smoke'].replace(0,np.nan)
df.head()
```

```python
df.loc[df['weight'] > 40, 'nobeyesdad'] = 'obese'
df.head()
```

## üõ† Grouping & Aggregation

```python
df.groupby('gender')['weight'].mean()
```

```python
df.groupby('gender')[['age', 'weight']].agg(['mean', 'std', 'count'])
```

## üõ† Index Operations

```python
df.index.name = 'id'
df.head()
```
